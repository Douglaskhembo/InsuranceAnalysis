```python
#Correlation analysis

Bmi_Age = Test[['bmi', 'age']]
Bmi_Charges = Test[['bmi', 'charges']]
Age_Charges = Test[['age', 'charges']]
Bmi_children = Test[['bmi', 'children']]

paper, axes = plt.subplots(2,2, figsize = (12,8))
Bmi_Age.plot(kind = 'hist', title = ('bmi and age distribution'), alpha = 0.7, ax = axes[0,0])
Bmi_Charges.plot(kind = 'hist', title = ('bmi and charges distribution'), alpha = 0.7, ax = axes[0,1])
Age_Charges.plot(kind = 'hist', title = ('Age and cahrge distribution'), alpha = 0.7, ax = axes[1,0])
Bmi_children.plot(kind = 'hist', title = ('bmi and children distribution'), alpha = 0.7, ax = axes[1,1])
plt.show()
```

```python
paper, axes = plt.subplots(2,2, figsize = (15,10))
sns.histplot(data = Test.age, ax=axes[0,0], kde = True).set(title = 'Age disribution')
sns.histplot(data = Test.bmi, kde = True, color = 'grey', ax = axes[0,1]).set(title = 'Bmi distribution')
sns.histplot(data = Test.children, kde = True, ax = axes[1,0]).set(title = 'Distibution of number of childreen')
sns.histplot(data = Test.charges, kde = True, color = 'red', ax = axes[1,1]).set(title = 'Charge distribution')
plt.suptitle('Skewness Distribution Analysis',size=20)
plt.show()
```

```python
Charge = pd.cut(Test['charges'], [0,5000,15000,25000, 35000, 45000, 55000, 65000])
Test.pivot_table('smoker', [Charge], aggfunc = sum)
```

We categorised charges into 4 groups, this is essential as it makes easy for us to analyse how charges fluctuate depending on whether an individual is a smoker or not

```python
sns.countplot(data = Test, x = Charge, hue = 'smoker')
plt.xticks(rotation = 90)
plt.show()
```

Insurance charges are higher for persons who smoke compared to those who don't. 

```python
Age = pd.cut(Test['age'], [0, 18, 30, 40, 45, 65])
Test.pivot_table('charges', [Charge], aggfunc = sum)
```

```python
sns.barplot(data = Test, x = Age, y = 'charges')
plt.show()
```

Individuals who are 18 years and below pay less insurance charges while while those of age category (45,45) pay the highest insurance charges
